```python
# Experiment 9: 0/1 Knapsack using Dynamic Programming

def knapsack_01(weights, values, capacity):
    n = len(values)

    # Create DP table
    dp = [[0 for x in range(capacity + 1)] for i in range(n + 1)]

    # Build table dp[][] in bottom-up manner
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(
                    values[i - 1] + dp[i - 1][w - weights[i - 1]],
                    dp[i - 1][w]
                )
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]


# Driver Code
weights = [2, 3, 4, 5]
values = [3, 4, 5, 6]
capacity = 5

max_value = knapsack_01(weights, values, capacity)
print("Maximum value in Knapsack =", max_value)
```

```python
# Experiment 10: Matrix Chain Multiplication using Dynamic Programming

import sys

def matrix_chain_order(p, n):
    # Create DP table
    dp = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            dp[i][j] = sys.maxsize

            for k in range(i, j):
                cost = dp[i][k] + dp[k + 1][j] + p[i - 1] * p[k] * p[j]
                dp[i][j] = min(dp[i][j], cost)

    return dp[1][n - 1]


# Driver Code
p = [1, 2, 3, 4]
n = len(p)

min_cost = matrix_chain_order(p, n)
print("Minimum number of multiplications is:", min_cost)
```

```python
# Lab Experiment 3: Exponential (Power) Function

# 1. Iterative (Naive) Approach
def power_iterative(x, n):
    result = 1
    for i in range(n):
        result = result * x
    return result


# 2. Divide and Conquer Approach - O(N)
def power_divide_conquer_on(x, n):
    if n == 0:
        return 1
    return x * power_divide_conquer_on(x, n - 1)


# 3. Divide and Conquer Approach - O(log N)
def power_divide_conquer_log(x, n):
    if n == 0:
        return 1

    temp = power_divide_conquer_log(x, n // 2)

    if n % 2 == 0:
        return temp * temp
    else:
        return x * temp * temp


# Driver Code
x = 2
n = 5

print("Iterative (Naive) Approach:", power_iterative(x, n))
print("Divide and Conquer O(N):", power_divide_conquer_on(x, n))
print("Divide and Conquer O(log N):", power_divide_conquer_log(x, n))
```
